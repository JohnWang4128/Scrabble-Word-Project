```python
from collections import Counter
import pandas as pd
file_path = "/NSWL2023.txt"


```

```python
letters = "tabind"
letter_counts = Counter(letters)

def can_form_word(word, letter_counts):
    word_counts = Counter(word.lower())
    for letter, count in word_counts.items():
        if count > letter_counts.get(letter, 0):
            return False
    return True
with open(file_path, "r") as file:
    words = [line.split()[0] for line in file.readlines()]

valid_words = sorted([word for word in words if can_form_word(word, letter_counts)])

df = pd.DataFrame(valid_words, columns=["Valid Words"])
display(df)
```
